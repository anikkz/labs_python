4. Преобразовать индексы объекта Series в столбец DataFrame

```python
a = pd.Series(np.arange(5), index = ['A', 'B', 'C', 'D', 'E'])
print(a)
f = pd.DataFrame(a.reset_index())
f.columns = ['a.index', 'a.values']
print(f)
```
