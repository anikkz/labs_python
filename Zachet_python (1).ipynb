Идет по улице прохожий, смотрит - на газоне двое рабочих.
Первый выкапывает яму, ждут, второй закапывает. Перекур. Потом все повторяется.
Прохожий смотрел-смотрел, не выдержал: "Мужики, а что вы делаете-то?"
-"Ну понимаешь, вообще мы деревья сажаем.
Я выкапываю яму, Петрович кладет туда саженец, Серега закапывает.
А сегодня Петрович заболел..."

Написать 2 варианта программы (программа должна запускаться):
Петрович здоров и сажает деревья;
Петрович заболел. Использовать цикл while. Выход из цикла - по окончании рабочего дня. Сколько у рабочих было саженцев - мы не знаем.

1. Выясняем состояние Петровича

```python
print("Петрович болеет? 1 - да / 0 - нет")
sostoyanie_petrovicha = int(input())
```

2. Объявляем начало рабочего дня

```python
print("8:00 - Начало рабочего дня")
```

3. Определяем порядок работы рабочих

```python
dict={}
time = 8

while time < 17:
    time+=0.5
    dict[time]="Выкопана яма"
    print(f'Время: {int(time)} часов, {int((time-int(time))*60)} минут', dict[time])
    if sostoyanie_petrovicha==0:
        time+=0.05
        dict[time]="Посажен саженец"
        
        print(f'Время: {int(time)} часов, {int((time-int(time))*60)} минут', dict[time])
        time+=0.35
        dict[time]="Закопана ямa"
        print(f'Время: {int(time)} часов, {int((time-int(time))*60)} минут', dict[time])
    elif sostoyanie_petrovicha==1:
        time+=0.5
        dict[time]="Закопана ямa"
        print(f'Время: {int(time)} часов, {int((time-int(time))*60)} минут', dict[time])
```

4. Объявляем конец рабочего дня
